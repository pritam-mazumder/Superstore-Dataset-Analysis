### **📝 Fill the null values in the dataset.**

```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(), inplace=True)
```

### **📝 Order priorities analysis**

```python
df['Order Priority'].value_counts()
```

The 'Critical' order priority has been duplicated twice. This needs to be resolved.

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'] = df['Order Priority'].replace('Critical ', 'Critical')
```

```python
df['Order Priority'].unique()
```

```python
plt.title('Count of Order Priority')
sns.countplot(data=df, x=df['Order Priority'])
plt.show()
```

***Conclusion:** Based on the analysis of the chart above, it is evident that all order priorities exceed 1750 in count. Specifically, 'high' priority exhibits the highest count, followed by 'low', 'not specified', and 'medium' priorities, while 'critical' priority registers the lowest count.*

*This distribution suggests that most orders are considered urgent or important, with fewer orders deemed as low priority or not specified.*

### **📝 Ship mode analysis**

```python
df['Ship Mode'].value_counts()
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values

plt.title('Count of Ship Mode')
plt.pie(y, labels=x, autopct='%.2f%%')
plt.legend(loc=1)
plt.show()
```

***Conclusion:** Based on the analysis of the chart above, it is evident that the count of 'Regular Air' is the predominant shipping mode, accounting for nearly three-quarters of all shipments at 74.64%. 'Delivery Truck' and 'Express Air' are used significantly less, at 13.61% and 11.74% respectively.*

*This suggests that the majority of items are shipped using regular air services, which could be due to a balance of cost and speed that suits most customers’ needs.*

### **📝 Product category analysis with respect to shipping mode**

```python
plt.title('Product category with respect to shipping mode')
sns.countplot(data=df, x='Ship Mode', hue='Product Category')
plt.show()
```

***Conclusion:** Based on the chart above, it is evident that 'Office Supplies' emerge as the most frequently shipped product category across all shipping modes, with the highest occurrence in 'Regular Air' shipping. 'Technology' products exhibit a lower shipping frequency compared to 'Office Supplies', while 'Furniture' shows the lowest shipping count in all modes.*

*This pattern suggests that 'Office Supplies' are likely in high demand or are commonly ordered, and that 'Regular Air' is preferred for its balance of speed and cost effectiveness.*

### **📝 Sub-product category with respect to product category sales analysis**

```python
df['Product Category'].unique()
```

```python
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.title('Sales of Office Supplies')
sns.countplot(data=df[df['Product Category']=='Office Supplies'], x='Product Category', hue='Product Sub-Category')

plt.subplot(1,3,2)
plt.title('Sales of Technology')
sns.countplot(data=df[df['Product Category']=='Technology'], x='Product Category', hue='Product Sub-Category')

plt.subplot(1,3,3)
plt.title('Sales of Furniture')
sns.countplot(data=df[df['Product Category']=='Furniture'], x='Product Category', hue='Product Sub-Category')
```

***Conclusion:** The above chart is a comparison of sales for office supplies, technology, and furniture, broken down by sub-category. Here are some key findings that can draw from the image:*

- *Office supplies is the top-selling category: Sales of office supplies are higher than both technology and furniture. Within the office supplies category, paper, binders and binder accessories and pens and art supplies are the top-selling sub-categories.*

- *Technology sales are in the middle: Technology sales are lower than office supplies but higher than furniture. Telephones and communication, and computer peripherals are the top-selling sub-categories within technology.*

- *Furniture sales are the lowest: Sales of furniture are the lowest compared to the other two categories. Office furnishing and chairs & chairmats are the top-selling sub-categories within furniture.*

*This highlights a potential opportunity for the business to strengthen the Furniture category or to further capitalize on the robust sales of Office Supplies and Technology.*
